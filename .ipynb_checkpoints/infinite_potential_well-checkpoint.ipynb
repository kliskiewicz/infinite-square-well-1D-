```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_bvp
```

```python
n = 3
def system_derivative(w, system):
    w, p = system
    return [p, -pow(n,2)*w]
```

```python
def bc(w_condition_1, w_condition_2):
    return np.array([w_condition_1[0], w_condition_2[0]])
```

```python
steps = 1000
space = np.linspace(0, np.pi, steps)
guess = np.ones((2, steps))*-1/2

for i in range(1, n+1):
    n = i
    sol = solve_bvp(system_derivative, bc, space, guess)
    plt.plot(sol.x, sol.y[0])
    print("n = %i" % i)
    plt.show()
```
